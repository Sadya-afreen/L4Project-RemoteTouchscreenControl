```python
import matplotlib.pyplot as plt
import pandas as pd
import pingouin as pg
import seaborn as sns

# Read experiment data into a data frame
data = pd.read_csv("P1-15Task3Data_Final.csv")

# Get rid of excess columns - we don't need these for this analysis
data = data.drop(columns=["Block","Task", "EulerA_Initial.tx", "EulerA_Initial.ty", "EulerA_Initial.tz","EulerA1.tx",
                          "EulerA1.ty", "EulerA1.tz","Cube1AngleofRot",
                          "Start Time","End Time","Difftime"])
data = data.groupby(['Participant', 'Condition'], as_index=False).agg({'timems': 'mean'})
tr  = data.loc[data['Condition'] == 'Smart Trackpad']
tb  = data.loc[data['Condition'] == 'Smart Trackball']
tp = data.loc[data['Condition'] == 'Smart Tilt Pointer']
rp = data.loc[data['Condition'] == 'Smart Rotate Pointer']

plot, axis = plt.subplots(figsize=(8, 4))
sns.distplot(tr['timems'],  hist=False, label="Trackpad",        color="orange")
sns.distplot(tb['timems'],  hist=False, label="Trackball",         color="red")
sns.distplot(tp['timems'], hist=False, label="Tilt Pointer", color="teal")
sns.distplot(rp['timems'], hist=False, label="Rotate Pointer", color="blue")
axis.set_xlabel("Time (ms)")
axis.set_ylabel("Density")
axis.legend()
```
