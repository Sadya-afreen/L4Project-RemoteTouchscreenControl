```python
import matplotlib.pyplot as plt
import pandas as pd
import pingouin as pg
import seaborn as sns
import scikit_posthocs as sp
# Read experiment data into a data frame
data = pd.read_csv("P1-15Rank_Data.csv")

# Get rid of excess columns - we don't need these for this analysis
#data = data.drop(columns=["Block","Task", "Target.x", "Target.y", "Targetsize.b","Targetsize.l"])
data = data.groupby(['Participant', 'Condition'], as_index=False).agg({'Score': 'mean'})
plot, axis = plt.subplots(figsize=(8, 4))
axis.set_title("Order of Preference (1-most preferred)")

sns.barplot(y='Score',x='Condition',data=data,errcolor = 'black', errwidth = 2,capsize=0.25,ci=95,color=sns.set_palette("ocean_r"))
axis.set_xlabel("Technique")
axis.set_ylabel("Rank")
ph_err_count = pg.friedman(dv='Score', within='Condition', subject='Participant', data=data)
pg.print_table(ph_err_count)
sp.posthoc_nemenyi(data,val_col='Score',group_col='Condition')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import pingouin as pg
import seaborn as sns
import scikit_posthocs as sp
# Read experiment data into a data frame
data = pd.read_csv("P1-15Rank_Data.csv")

# Get rid of excess columns - we don't need these for this analysis
#data = data.drop(columns=["Block","Task", "Target.x", "Target.y", "Targetsize.b","Targetsize.l"])
data = data.groupby(['Participant', 'Condition'], as_index=False).agg({'overall_workload': 'mean'})
plot, axis = plt.subplots(figsize=(8, 4))
axis.set_title("Mean Workload")
sns.barplot(y='overall_workload',x='Condition',data=data,errcolor = 'black', errwidth = 2,capsize=0.25,ci=95,color=sns.set_palette("ocean_r"))
axis.set_xlabel("Technique")
axis.set_ylabel("Workload")

tr  = data.loc[data['Condition'] == 'Smart Trackpad']
tb  = data.loc[data['Condition'] == 'Smart Trackball']
tp = data.loc[data['Condition'] == 'Smart Tilt Pointer']
rp = data.loc[data['Condition'] == 'Smart Rotate Pointer']
print(tb)
plot, axis = plt.subplots(figsize=(8, 4))
sns.distplot(tr['overall_workload'],  hist=False, label="Trackpad",        color="orange")
sns.distplot(tb['overall_workload'],  hist=False, label="Trackball",         color="red")
sns.distplot(tp['overall_workload'], hist=False, label="Tilt Pointer", color="teal")
sns.distplot(rp['overall_workload'], hist=False, label="Rotate Pointer", color="blue")
axis.set_xlabel("overall_workload")
axis.set_ylabel("Density")
axis.legend()

aov_workload = pg.rm_anova(dv='overall_workload', within='Condition', subject='Participant', data=data)
pg.print_table(aov_workload)

ph_workload = pg.pairwise_ttests(dv='overall_workload', within='Condition', subject='Participant', data=data)
pg.print_table(ph_workload)
```
