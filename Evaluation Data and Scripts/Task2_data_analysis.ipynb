```python
import matplotlib.pyplot as plt
import pandas as pd
import pingouin as pg
import seaborn as sns

# Read experiment data into a data frame
data = pd.read_csv("P1-15Task2Data.csv")

# Get rid of excess columns - we don't need these for this analysis
data = data.drop(columns=["Block","Task", "temp.y", "temp.z","final.y","final.z"])
data = data.groupby(['Participant','Condition'], as_index=False).agg({'Euclidean_distance': 'mean'})
#data=data.loc[data['Condition']]
#print(data)
tr  = data.loc[data['Condition'] == 'Smart Trackpad']
tb  = data.loc[data['Condition'] == 'Smart Trackball']
tp = data.loc[data['Condition'] == 'Smart Tilt Pointer']
rp = data.loc[data['Condition'] == 'Smart Rotate Pointer']

plot, axis = plt.subplots(figsize=(8, 4))
sns.barplot(x="Condition", y="Euclidean_distance",data=data,errcolor = 'black', errwidth = 2,capsize=0.25,ci=95 )
axis.set_title("Mean Euclidean Distance")
axis.set_xlabel("Technique")
axis.set_ylabel("Distance")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import pingouin as pg
import seaborn as sns

# Read experiment data into a data frame
data = pd.read_csv("P1-15Task2Data.csv")

# Get rid of excess columns - we don't need these for this analysis
data = data.drop(columns=["Block","Task", "temp.y", "temp.z","final.y","final.z"])
data = data.groupby(['Participant', 'Condition'], as_index=False).agg({'timems': 'mean'})
aov_time = pg.rm_anova(dv='timems', within='Condition', subject='Participant', data=data)
pg.print_table(aov_time)
```
